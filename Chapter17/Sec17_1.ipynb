```python
import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the MNIST data
mnist = fetch_openml('mnist_784', parser='auto')
X = mnist.data / 255.0  # Scale the data to [0, 1]
y = mnist.target

# Apply PCA to reduce the dimensionality of the data
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)
X_pca = np.array(X_pca)
y_pca = np.array(y)


```

```python
# Create the 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca[:500, 0], X_pca[:500, 1], X_pca[:500, 2], c=y_pca[:500].astype(int), s=10)


#plt.savefig('MNIST1.eps', format='eps', bbox_inches='tight')
plt.show()
```

```python

n1=1 # first number
n2=4 # second number
indices_with_n1 = np.where(y.astype(int) == n1)[0]
indices_with_n2 = np.where(y.astype(int) == n2)[0]

indices_n1n2 = np.concatenate((indices_with_n1[:1000], indices_with_n2[:1000]))
label= y[indices_n1n2]


cmap = 'coolwarm'  # Choose a colormap, e.g., 'viridis', 'plasma', 'inferno'
colors = label[indices_n1n2].astype(int)  # Get the color values based on the labels


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca[indices_n1n2, 0], X_pca[indices_n1n2, 1], X_pca[indices_n1n2, 2], c=colors, cmap=cmap, s=10)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
#ax.set_zlabel('PC3')
fig.text(0.80, 0.75, 'PC3')

plt.savefig('MNIST2.eps', format='eps', bbox_inches='tight')
plt.show()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca[indices_n1n2, 0], X_pca[indices_n1n2, 1], X_pca[indices_n1n2, 2], s=10)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
#ax.set_zlabel('PC3')
fig.text(0.80, 0.75, 'PC3')

#plt.savefig('MNIST3.eps', format='eps', bbox_inches='tight')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=5, ncols=5, figsize=(8, 8))
axs = axs.ravel()
Xs = np.array(X)

for i in range(25):
    sample = Xs[i,:]
    axs[i].imshow(sample.reshape(28, 28), cmap='gray')
    axs[i].set_yticks([])
    axs[i].set_xticks([])

plt.tight_layout()

plt.savefig('MNIST4.eps', format='eps', bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_openml

# Load the MNIST data
mnist = fetch_openml('mnist_784', parser='auto')
X, y = mnist.data, mnist.target
X = X / 255.0 # Scale the pixel values to [0, 1]

# Perform PCA on the data
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)

# Create the 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=mnist.target.astype(int), s=20)

# Set the labels and limits of the plot
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
#ax.set_zlabel('PC3')
fig.text(0.80, 0.75, 'PC3')

plt.savefig('MNIST9.eps', format='eps', bbox_inches='tight')

plt.show()
```
