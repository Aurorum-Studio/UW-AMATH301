```python
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt
from numpy import linalg 

results=loadmat('yalefaces.mat')
X=results['X']
print(X.shape)

fig = plt.figure(figsize=(18, 18))
for k in range(9):
    Xp = X[:, k].reshape((32, 32), order='F')
    uplot = Xp.T
    ax = fig.add_subplot(3, 3, k+1)
    ax.imshow(uplot,cmap='gray')
    plt.xticks([]); plt.yticks([])
```

```python
m=10
Xm=X[:, 0:9]
C=np.matmul(Xm.T,Xm)
print(Xm.shape)
print(C.shape)
print(C)
plt.pcolor(C,cmap='gray')
```

```python
C=np.matmul(X,X.T)
D,V = np.linalg.eig(C)

sorted_indices = np.argsort(np.abs(D))[::-1] 
Dsort = D[sorted_indices]
Vsort =V[:, sorted_indices]

fig = plt.figure(figsize=(18, 18))
for k in range(9):
    Up = Vsort[:, k].reshape((32, 32), order='F')
    uplot = Up.T
    ax = fig.add_subplot(3, 3, k+1)
    ax.imshow(uplot,cmap='gray')
    plt.xticks([]); plt.yticks([])

```

```python
plt.stem(Dsort, markerfmt='ro')
plt.xlim(-2, 20)
plt.show()
```

```python
u,s,v=np.linalg.svd(X)
fig = plt.figure(figsize=(15, 10))
for k in range(9):
    Up = u[:, k].reshape((32, 32), order='F')
    uplot = Up.T
    ax = fig.add_subplot(3, 3, k+1)
    ax.imshow(uplot,cmap='gray')

print(s.shape)
print(s[0:5])
print(v.shape)
print(np.sum(s))
```

```python
plt.stem(s, markerfmt='ro')
plt.xlim(-2, 20)
plt.show()
```

```python
import numpy as np
from scipy.linalg import eig

# Generate a random 20x20 matrix
matrix = np.random.rand(20, 20)

# Compute the first 5 eigenvalues and eigenvectors
num_eigenvalues = 5
eigenvalues, eigenvectors = linalg.eig(matrix, k=num_eigenvalues)

print("First", num_eigenvalues, "eigenvalues:")
print(eigenvalues)
print("\nCorresponding eigenvectors:")
print(eigenvectors)
```
