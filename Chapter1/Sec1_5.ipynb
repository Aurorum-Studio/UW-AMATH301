Section 1.1:  Vectors and Matrices

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10,10.1,0.1)
y = np.sin(x)

x2 = [-5,np.sqrt(3),np.pi,4]
y2 = np.sin(x2)

x3 = np.linspace(-10,10,64)
y3 = x3*np.sin(x3)

plt.plot(x,y)
plt.plot(x2,y2)
plt.plot(x3,y3)
```

```python
# or you can plot all in one line

#plt.plot(x,y,x2,y2,x3,y3)
plt.plot(x,y,x2,y2,'g*',x3,y3,'mo:')
```

```python
# add labels and legends

plt.plot(x,y,label='Data Set 1')
plt.plot(x2,y2,label='Data Set 2')
plt.plot(x3,y3,label='Data Set 3')

plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Example Graph')
plt.legend()
```

```python
# Or alternatively

fig = plt.subplot(311)
plt.plot(x,y), plt.axis([-10, 10, -10, 10])
a=5
plt.title('value of a=' + str(a))

fig = plt.subplot(312)
plt.plot(x2,y2,'g*'), plt.axis([-10, 10, -10, 10])

fig = plt.subplot(313)
plt.plot(x3,y3,'mo:')


```
