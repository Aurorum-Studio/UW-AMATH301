Section 13.7:  Spectrograms and the Gabor Transform

```python
import numpy as np
from numpy.fft import fft, fftshift, ifft

L = 10                          # time slot
n = np.power(2,11)              # number Fourier modes
t = np.linspace(0, L, n+1)
t = t[:-1]                      # time domain
k = (2*np.pi/(L))*np.concatenate((np.arange(0,n/2), np.arange(-n/2,0)))  # frequency domain
ks = fftshift(k)
```

```python
# create signal
S = (3*np.sin(2*t) + 0.5*np.sinh(0.5*(t-3))/np.cosh(0.5*(t-3)) + 0.2*np.exp(-(t-4)**2) + 1.5*np.sin(5*t) + 4*np.cos(3*(t-6)**2))/10 + (t/20)**3
St = fft(S)
 
import matplotlib.pyplot as plt    

fig1, (sig,sigt) = plt.subplots(2,1)
plt.rcParams['figure.figsize'] = [10, 10]

sig.plot(t, S, 'k')
sig.set_title('Time Signal')

sigt.plot(k,abs(St),'k')
sigt.set_xlim(-50, 50)
sigt.set_title('Fourier Domaina')
```

Gabor Window:  Gaussians with varying window widths

```python
width=(10, 1, 0.2)

plt.rcParams['figure.figsize'] = [10, 10]
fig=plt.figure()

for j in np.arange(0,len(width)):
    g = np.exp(-width[j]*(t-4)**2)
    
    ax = fig.add_subplot(311+j)
    ax.plot(t,S,'r',t,g,'k')
    ax.set_ylabel('S, g')
    ax.set_xlabel('time')
    
```

Pull a signal out of a window

```python
plt.rcParams['figure.figsize'] = [10, 10]
fig=plt.figure()

g = np.exp(-2*(t-4)**2)
Sg=g*S
Sgt=fftshift(fft(Sg))

ax = fig.add_subplot(311)
ax.plot(t,S,'r',t,g,'k')

ax = fig.add_subplot(312)
ax.plot(t,Sg,'k')

ax = fig.add_subplot(313)
ax.plot(ks,abs(Sgt),'k')
ax.set_xlim(-50, 50)

```

Sliding Gabor Window

```python
plt.rcParams['figure.figsize'] = [10, 10]
fig=plt.figure()

dt=.1
tslide = np.arange(0,10+dt,dt)

for j in np.arange(0,len(tslide)):
    g = np.exp(-1*(t-tslide[j])**2)
    Sg=g*S
    Sgt=fftshift(fft(Sg))

    ax = fig.add_subplot(311)
    ax.plot(t,S,'r',t,g,'k')

    ax = fig.add_subplot(312)
    ax.plot(t,Sg,'k')

    ax = fig.add_subplot(313)
    ax.plot(ks,abs(Sgt),'k')
    ax.set_xlim(-100, 100)
    
    plt.pause(0.1)
    plt.show()
    
```
